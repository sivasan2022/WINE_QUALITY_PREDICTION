WINE QUALITY PREDICTION USING LINEAR REGRESSION MODEL

Importing necessary libraries

```python
import pandas as pd
df=pd.read_csv('WineQT.csv')
df
```

Splitting x and y variables

```python
x=df.iloc[:,:-2]
x
```

```python
y=df.iloc[:,-2]
y
```

Splitting the dataset for testing and training

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
ytrain.shape
```

```python
ytest.shape
```

Building a linear regression model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
model = LinearRegression()
# Train the model on the training data
model.fit(xtrain, ytrain)

# Make predictions on the test data
y_pred = model.predict(xtest)
```

Performing a sample prediction

```python
y_pred = model.predict([[7.8,0.760,0.04,2.3,0.092,15.0,54.0,0.99700,3.26,0.65,9.8]])
y_pred
```

Evaluating the model

```python
# Evaluate the model
mse = mean_squared_error(ytest, y_pred)
r2 = r2_score(ytest, y_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

To improve the model, data preprocessing is performed using LabelEncoder

DATA PREPROCESSING

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
normalized_data=df.apply(LabelEncoder().fit_transform)
normalized_data.head()
```

```python
normalized_data.dtypes
```

Again splitting the dataset for x and y

```python
x1=normalized_data.iloc[:,:-2]
x1
```

```python
y1=normalized_data.iloc[:,-2]
y1
```

Splitting the dataset

```python
from sklearn.model_selection import train_test_split
x1train,x1test,y1train,y1test=train_test_split(x1,y1,test_size=0.2)
```

Building Linear Regression model using normalized data

```python
model = LinearRegression()
# Train the model on the training data
model.fit(x1train, y1train)

# Make predictions on the test data
y1_pred = model.predict(x1test)
```

Evaluating the model

```python
# Evaluate the model
mse = mean_squared_error(y1test, y1_pred)
r2 = r2_score(y1test, y1_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)
```
